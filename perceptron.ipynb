```python
# AND gate
# Define initial weights, initial learning rate, inputs, and target outputs
w1 = 0
w2 = 0
b = -1
lr = 0.5

x1 = [0,0,1,1]
x2 = [0,1,0,1]
target = [0,0,0,1]

count = 0 # tracks num of correct entries
```

```python
count = 0
while count < 4:
  for i in range(4):
    print(f"Inputs: x1 = {x1[i]}, x2 = {x2[i]}")
    z = (x1[i]*w1)+(x2[i]*w2)+b
    if z > 0:
      predicted = 1
    else:
      predicted = 0
    err = target[i]-predicted
    if err == 0:
      count += 1
      print(f"One set of inputs satisfied, count = {count}")
      if count == 4: break
    else:
      delta_w1 = lr*err*x1[i]
      delta_w2 = lr*err*x2[i]
      w1 = delta_w1+w1
      w2 = delta_w2+w2
      b = lr*err*1
      print(f"Error = {err}, Weights Updated: w1 = {w1}, w2 = {w2}, b = {b}")
      count = 0
print(f"\nFinal weights: w1 = {w1}, w2 = {w2}, b = {b}")
```
